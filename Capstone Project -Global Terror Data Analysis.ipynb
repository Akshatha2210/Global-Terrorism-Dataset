#### Chart - 1

````python
# Chart - 1 visualization code
#Analysis of the number of attacks per year
f = plt.figure(figsize=(25, 5))
sns.set(font_scale = 1.1)
sns.set_theme(style = "darkgrid")
xaxis = sns.countplot(x = 'iyear', data = terror_df)
xaxis.set_xticklabels(xaxis.get_xticklabels(), rotation=60)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Year', fontsize=12)
plt.title('Number of Terrorist Attack by Year', fontsize = 12)
````

##### 1. Why did you pick the specific chart?

I provided the code for a countplot because it's a suitable choice for visualizing the distribution of a categorical variable, in this case, the number of terrorist attacks over different years. Here's why the countplot was chosen:

Categorical Data: The variable being plotted (iyear, which likely represents the year of the terrorist attack) is categorical, as it represents distinct categories (years) rather than continuous values.

Frequency Distribution: The countplot is specifically designed to show the frequency distribution of categorical data. It displays the count of observations in each category (year) using bars.

Ease of Interpretation: Countplots are easy to interpret and provide a clear visual representation of the distribution of categorical data. In this case, it allows for a quick assessment of the number of terrorist attacks recorded for each year.

Readability: By rotating the x-axis labels, we ensure that the year labels don't overlap and remain readable, even with a large number of years.

Aesthetics: The choice of using a dark grid theme (style="darkgrid") can enhance the visual appeal of the plot, making it easier to distinguish between bars.

Overall, the countplot is a suitable choice for visualizing the distribution of terrorist attacks over different years, making it easier to identify any trends or patterns over time.

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
numerical_columns = terror_df.select_dtypes(include='number')

# Calculating correlation matrix
correlation_matrix = numerical_columns.corr()

# Plotting correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
````

##### 1. Why did you pick the specific chart?

The specific chart, a correlation heatmap, was chosen to visualize the pairwise correlations between numerical variables in the dataset for several reasons:

1. **Comprehensive Visualization:** A correlation heatmap provides a comprehensive overview of the relationships between all pairs of numerical variables in the dataset. This allows for a holistic understanding of how variables are related to each other, helping to identify patterns and dependencies.

2. **Ease of Interpretation:** The heatmap format makes it easy to interpret the strength and direction of correlations visually. Warm colors (e.g., red) represent positive correlations, while cool colors (e.g., blue) represent negative correlations. The intensity of the color indicates the strength of the correlation, allowing for quick identification of significant relationships.

3. **Identifying Multicollinearity:** By visualizing correlations, the heatmap helps identify multicollinearity, which occurs when independent variables are highly correlated with each other. Multicollinearity can affect the performance of regression models and other statistical analyses, so detecting and addressing it is important for accurate data analysis.

4. **Efficient Exploration:** The heatmap allows for efficient exploration of the entire correlation matrix, especially in datasets with numerous variables. Instead of examining each correlation coefficient individually, the heatmap provides a clear and concise summary of all correlations, facilitating quick insights into the data.

5. **Readability:** The heatmap format ensures that correlations are presented in a visually appealing and easily interpretable manner. The inclusion of annotation (actual correlation values) within each cell further enhances readability, allowing for precise interpretation of correlation coefficients.

Overall, the correlation heatmap is a suitable choice for exploring relationships between numerical variables in the dataset, providing valuable insights into the underlying structure and interdependencies within the data.

##### 2. What is/are the insight(s) found from the chart?

The insights derived from the correlation heatmap can provide valuable information about the relationships between numerical variables in the dataset:

Strong correlations indicate a high degree of linear relationship between variables, while weak correlations suggest little to no linear relationship.
